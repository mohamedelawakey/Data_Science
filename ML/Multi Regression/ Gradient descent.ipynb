# Steps 

1. start with random parameters or zero like linear regression slope = 0, intercept = 0
2. prediction as y` = slope * x + intercept
3. cost function like mean square error -> MSE = J(slope, intercept) = (1 / n) * sum(y - y`)^2
4. Grediants:
    - J∂ / slope∂ = -(2 / n) * sum(x * (y - y`))
    - J∂ / intercept∂ = -(2 / n) * sum(y - y`)
5. update the steps:
    - slope = slope - learning rate * J∂ / slope∂
    - intercept = intercept - learning rate * J∂ / intercept∂
6. iterations

```python
import numpy as np
```

```python
def GrediantDesent(x, y):
    slope, intercept = 0, 0
    learning_rate = 0.08
    iterations = 10000
    n = len(x)

    for i in range(iterations):
        y_prediction = slope * x + intercept
        cost = (1 / n) * sum([val ** 2 for val in (y - y_prediction)])
        Grediants_slope = -(2 / n) * sum(x * (y - y_prediction))
        Grediants_intercept = -(2 / n) * sum((y - y_prediction))
        slope = slope - learning_rate * Grediants_slope
        intercept = intercept - learning_rate * Grediants_slope
        print("m {}, b {}, cost {} iteration {}". format(slope, intercept, cost, i))
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

GrediantDesent(x,y)
```
